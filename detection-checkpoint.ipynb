```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import keras
from tensorflow.keras.preprocessing.image import img_to_array
from keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing.image import load_img
from keras.applications.vgg19 import VGG19, preprocess_input #decode_prediction
```

```python
from tensorflow.keras.optimizers.legacy import Adam
```

```python
train_datagen = ImageDataGenerator(zoom_range= 0.5, shear_range= 0.3, rescale=1/255, horizontal_flip=True)
val_datagen = ImageDataGenerator(rescale=1/255)
```

```python
train = train_datagen.flow_from_directory(directory= 'C:\\Users\\aerin\\Downloads\\archive (1)\\PlantVillage', target_size= (256,256), batch_size=31)
```

```python
val = train_datagen.flow_from_directory(directory= 'C:\\Users\\aerin\\Downloads\\archive (1)\\PlantVillage', target_size= (256,256), batch_size=31)
```

```python
t_img, label = train.next()
```

```python
def plotImage(img_arr, label):
    for im, l in zip(img_arr, label):
        plt.figure(figsize=(5,5))
        plt.imshow(im)
        plt.show()
```

```python
plotImage(t_img[:3], label[:3])
```

```python
from keras.layers import Dense, Flatten
from keras.models import Model
from keras.applications.vgg19 import VGG19
import keras
```

```python
base_model = VGG19(input_shape=(256,256,3), include_top= False)
```

```python
for layer in base_model.layers:
    layer.trainable = False
```

```python
base_model.summary()
```

```python
x = Flatten()(base_model.output)
x = Dense(units= 4, activation='softmax')(x)
model = Model(base_model.input, x)
```

```python
model.summary()
```

```python
model.compile(optimizer= 'adam', loss= keras.losses.categorical_crossentropy, metrics=['accuracy'])
```

```python
from keras.callbacks import ModelCheckpoint, EarlyStopping
es = EarlyStopping(monitor= 'val_accuracy', min_delta= 0.01, patience= 3, verbose=1)
    mc = ModelCheckpoint(filepath="..\\archive (1)\\POTATO.h5", monitor= 'val_accuracy', min_delta= 0.01, patience= 3, verbose=1, save_best_only= True)
cb = [es, mc]
```

```python
his = model.fit(train, steps_per_epoch=16, epochs= 10, verbose= 1, callbacks= cb, validation_data= val, validation_steps= 16 )
```

```python
from keras.models import load_model
model = load_model('C:\\Users\\aerin\\Downloads\\archive (1)\\POTATO.h5')
```

```python
acc = model.evaluate(val)[1]
print(f"The accuracy of your model is = {acc*100} %")
```
